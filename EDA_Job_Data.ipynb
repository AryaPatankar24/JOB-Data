#18. Group by location and compute max salary.

```python
# Remove commas, currency symbols if any, and convert to numeric
df['salary'] = df['salary'].replace('[\$,₹,]', '', regex=True)  # Remove $, ₹
df['salary'] = df['salary'].replace(',', '', regex=True)        # Remove commas
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')     # Convert to float
```

```python
max_salary_by_location = df.groupby('location')['salary'].max().sort_values(ascending=False)
print(max_salary_by_location)
```

#27. Most common words in job descriptions.

```python
all_text = ' '.join(df['description'].dropna().astype(str).tolist())
```

```python
import re
from collections import Counter
```

```python
words= re.findall(r'\b\w+\b', all_text.lower())

word_counts = Counter(words)
```

```python
top_words= word_counts.most_common(20)
for word, count in top_words:
  print(f'{word}: {count}')
```

```python
top_df = pd.DataFrame(top_words, columns=['word', 'count'])

plt.figure(figsize=(10, 6))
sns.barplot(data=top_df, x='count', y='word', palette='mako')
plt.title('Top 20 Most Common Words in Job Descriptions')
plt.xlabel('Count')
plt.ylabel('Word')
plt.grid(True)
plt.tight_layout()
plt.show()
```
